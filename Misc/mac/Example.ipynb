```python
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.weight'] = 'bold'
```

```python
temp = [19.7, 19.6, 19.7, 19.8, 19.4]
time = [0,1,2,3,4]

plt.figure(figsize=(8,6))
plt.plot(time,temp,marker='o',color='g',lw=2)
plt.grid(True)
plt.ylim(19,20)
plt.xlim(0,5)
plt.xlabel('Time [unit]',weight='bold',fontsize=20)
plt.ylabel('Temp [unit]')
plt.title('Example plot')
plt.savefig('temp_v_time.png')
```
